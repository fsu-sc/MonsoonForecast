```python
import os
import numpy as np
#import xarray as xr
#import matplotlib.pyplot as plt

# Path to the folder containing NetCDF files
folder_path = '/Net/elnino/data/obs/ERA5/global/daily/'

# List all NetCDF files in the folder
nc_files = [f for f in os.listdir(folder_path) if f.endswith('.nc')]

# Create a dataset to store file sizes
sizes_dataset = xr.Dataset()

file_sizes = []
# Iterate through each NetCDF file
for nc_file in nc_files:
    file_path = os.path.join(folder_path, nc_file)
    file_size = os.path.getsize(file_path)
    sizes_dataset[nc_file] = file_size
    file_sizes.append(file_size)

#print("List of file sizes:", file_sizes)
print("How many files:", len(file_sizes))
print("Mean File size:", (np.mean(file_sizes) /  1073741824), 'GB')
print("File Size Range:", ((np.max(file_sizes) - np.min(file_sizes)) / 1073741824), 'GB')

# Plot the distribution of file sizes

#plt.line(nc_files, file_sizes)
#plt.xlabel('NetCDF Files')
#plt.ylabel('File Size (bytes)')
#plt.title('Distribution of NetCDF File Sizes')
#plt.xticks(rotation=45, ha='right')
#plt.tight_layout()
#plt.show()

# Create a histogram
plt.hist(file_sizes, bins=2, color='blue', edgecolor='black')

# Add labels and title
plt.xlabel('File Size')
plt.ylabel('Frequency')
plt.title('Distribution of File Sizes')
```

```python
#Mean precipitation for whole region for a year
#Note: this is global region; cropping for Florida may end up being: 24 to 33 deg lat and -78 to 88 deg long
import matplotlib.pyplot as plt

filepath_ex_tp = '/Net/elnino/data/obs/ERA5/global/daily/tp_era5_day_2002.nc'
ds_ex_tp = xr.open_dataset(filepath_ex_tp)

mean_tp = ds_ex_tp['tp'].mean(dim=('latitude', 'longitude'))
mean_tp.plot.line(x='time')
plt.title('2002 Time Series Plot of Mean Total Precipitation')
plt.xlabel('Time')
plt.ylabel('Mean Total Precipitation')
plt.show()
```

```python
#cropping for Florida may end up being: 24 to 33 deg lat and -78 to 88 deg long
#florida_region = ds_ex_tp.sel(latitude=slice(24, 33), longitude=slice(-88, -78))

#FLorida Crop
#ds_ex_tp = ds_ex_tp.sel(latitude=slice(24, 33), longitude=slice(92, 102))
ds_ex_tp = xr.open_dataset(filepath_ex_tp)
import cartopy.crs as ccrs

projection = ccrs.PlateCarree()

fig, ax = plt.subplots(subplot_kw={'projection': projection}, figsize=(10, 6))

ds_ex_tp_transposed = ds_ex_tp.transpose('time', 'latitude', 'longitude')
#ds_ex_tp_transposed['tp'].plot.pcolormesh(ax=ax, transform=projection, x='latitude', y='longitude')
ds_one_day = ds_ex_tp_transposed.sel(time='2002-01-01')

ds_one_day
ds_one_day['tp'].plot.pcolormesh(ax=ax, transform=projection, cmap='viridis', x='longitude', y='latitude', add_colorbar=True)

#cax = ds_transposed['var1'].isel(time=0).plot.imshow(ax=ax, transform=projection, cmap='viridis', add_colorbar=False)

ax.coastlines()
ax.gridlines(draw_labels=True, linestyle='--', color='black', alpha=0.5)

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

#plt.colorbar(cax, orientation='horizontal', label='tp')

plt.title('World Map Plot of mslp for Jan 1 2002')
plt.show()
```
